# Нейронная сеть

В данной работе используется набор данных о неисправностях стальных плит
(https://archive.ics.uci.edu/ml/machine-learning-databases/00198/Faults.NNA)

#### Разбиваем данные на обучающую и тестовую выборки:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 2)
```

## Обучаем классификатор

```python
from sklearn.neural_network import MLPClassifier
#from pylab import rcParams
#rcParams['figure.figsize'] = (15.0,8.0)
```

```python
mlp_model = MLPClassifier(hidden_layer_sizes = (100,), solver = 'lbfgs', 
                          activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)
```

#### Проверяем качество классификатора:

```python
y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
```

#### Находим ошибки на обучающей и тестовой выборках:

```python
print(mlp_model.score(X_train, y_train), mlp_model.score(X_test, y_test))
```

# Подбор гиперпараметров

#### Подберем оптимальное значение параметра регуляризации alpha:

```python
alpha_arr = np.array([10**i for i in range(5,-5,-1)])
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (16,),max_iter=1,
                              solver = 'lbfgs', activation = 'logistic')
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))  
```

#### Визуализируем на графике значения ошибок на тренировочной и тестовой выборке:

```python
plt.semilogx(alpha_arr, train_err, 'b-o', label = 'test')
plt.semilogx(alpha_arr, test_err, 'r-o', label = 'train')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass    
```

### Находим минимальное значение ошибки:

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

### Находим оптимальное значение alpha:

```python
alpha_opt = alpha_arr[test_err == min_test_err]
print(alpha_opt)
```

```python
a = test_err == min_test_err
```

```python
a
```

```python
alpha_opt = alpha_opt[0]
```

### Строим нейронную сеть при найденном оптимальном значении alpha

Повторим обучение и посмотрим на ошибки на тренировочной и тестовой выборке:

```python
mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (100,),
                          solver = 'lbfgs', activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)

y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print(np.mean(y_train != y_train_pred))
print(np.mean(y_test != y_test_pred))
```

## Матрица рассогласования:

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test_pred, y_test))
```

### Находим ошибки в работе нейронной сети:

```python
err_number = 0
for i in range(N_test):
    leaf = X_test[i:i+1]
    cls = y_test[i:i+1]
    cls_pred = mlp_model.predict(leaf)
    if cls_pred[0] != cls.values[0]:
        print("\nWrong class = ",cls_pred[0], "\nObject: ", leaf.head())
        err_number +=1
```
